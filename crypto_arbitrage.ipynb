### Calculate the Arbitrage Profits

```python
# Measure the arbitrage spread between the two exchanges

arbitrage_spread_early = coinbase_sliced.loc["2018-01-10"] - bitstamp_sliced.loc["2018-01-10"] 
arbitrage_spread_middle = coinbase_sliced.loc["2018-02-06"] - bitstamp_sliced.loc["2018-02-06"]
arbitrage_spread_late = coinbase_sliced.loc["2018-03-25"] - bitstamp_sliced.loc["2018-03-25"]

spread_early = arbitrage_spread_early[arbitrage_spread_early>0]
print(spread_early.describe())
spread_middle = arbitrage_spread_middle[arbitrage_spread_middle>0]
print(spread_middle.describe())
spread_late = arbitrage_spread_late[arbitrage_spread_late>0]
print(spread_late.describe())
```

```python
# Calculate the spread returns 
spread_return_early = spread_early / bitstamp_sliced.loc["2018-01-10"]
spread_return_middle = spread_middle / bitstamp_sliced.loc["2018-02-06"]
spread_return_late = spread_late / bitstamp_sliced.loc["2018-03-25"]

# Review the spread return DataFrame
print(spread_return_early.describe())
print(spread_return_middle.describe())
print(spread_return_late.describe())
```

```python
# Determine the number of times trades with positive returns 
# exceed the 1% minimum threshold (.01) that is needed to cover costs
profitable_trades_early = spread_return_early[spread_return_early > .01]
print(profitable_trades_early.head(5))
profitable_trades_middle = spread_return_middle[spread_return_middle > .01]
print(profitable_trades_middle.head(5))
profitable_trades_late = spread_return_late[spread_return_late > .01]
print(profitable_trades_late.head(5))

# Review the first five profitable trades
```

```python
# Generate the summary statistics for the profitable trades
print(profitable_trades_early.describe())
print(profitable_trades_middle.describe())
print(profitable_trades_late.describe())
```

```python
# Calculate the potential profit per trade in dollars 
# Multiply the profitable trades by the cost of the Bitcoin that was purchased
profit_early = profitable_trades_early * bitstamp_sliced.loc["2018-01-10"]
profit_early
profit_middle = profitable_trades_middle * bitstamp_sliced.loc["2018-02-06"]
profit_middle
profit_late = profitable_trades_late * bitstamp_sliced.loc["2018-03-25"]
profit_late

# Drop any missing values from the profit DataFrames
profit_per_trade_early = profit_early.dropna()
profit_per_trade_middle = profit_middle.dropna()
profit_per_trade_late = profit_late.dropna()

# View the DataFrames profit summary
print(profit_per_trade_early.describe())
print(profit_per_trade_middle.describe())
print(profit_per_trade_late.describe())
```

```python
# Plot the results for the profit per trade DataFrames

profit_per_trade_early.plot(
    figsize = (10, 5),
    title = "Arbitrage Profits",
    color = "green"
)

profit_per_trade_middle.plot(
    figsize = (10, 5),
    title = "Arbitrage Profits",
    color = "purple"
)

profit_per_trade_late.plot(
    figsize = (10, 5),
    title = "Arbitrage Profits",
    color = "orange"
)
```

```python
# Calculate the sum of the potential profits for the profit per trade DataFrames
profit_sum_early = profit_per_trade_early.sum()
profit_sum_middle = profit_per_trade_middle.sum()
profit_sum_late = profit_per_trade_late.sum()
```

```python
# Calculate the cumulative profits over time for the profit per trade DataFrames
cumulative_profit_early = profit_per_trade_early.cumsum()
cumulative_profit_middle = profit_per_trade_middle.cumsum()
cumulative_profit_late = profit_per_trade_late.cumsum()
```

```python
# Plot the cumulative sum of profits for the profit per trade DataFrames
cumulative_profit_early.plot(figsize=(10, 7), title="Cumulative Arbitrage Profits")
cumulative_profit_middle.plot(figsize=(10, 7), title="Cumulative Arbitrage Profits")
cumulative_profit_late.plot(figsize=(10, 7), title="Cumulative Arbitrage Profits")
```
